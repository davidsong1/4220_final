```python
seq_len = 228
num_seq = 4
f = SeqIO.parse(align, 'fasta')

with open('phylo_info.csv', 'w') as file:
    for x in range(228):
        seq = ""
        counter = 0
        informative = False
        f = SeqIO.parse(align, 'fasta')
        for row in f:
            seq += row.seq[x]
        if Seq(seq).count('A') >= 2:
            counter += 1
        if Seq(seq).count('C') >= 2:
            counter += 1
        if Seq(seq).count('T') >= 2:
            counter += 1
        if Seq(seq).count('G') >= 2:
            counter += 1
        if counter >= 2:
            informative = True
        file.write(str(x)+","+str(informative)+'\n')
```

```python
codon_table = {
        'A': ('GCT', 'GCC', 'GCA', 'GCG'),
        'C': ('TGT', 'TGC'),
        'D': ('GAT', 'GAC'),
        'E': ('GAA', 'GAG'),
        'F': ('TTT', 'TTC'),
        'G': ('GGT', 'GGC', 'GGA', 'GGG'),
        'H': ('CAT', 'CAC'),
        'I': ('ATT', 'ATC', 'ATA'),
        'K': ('AAA', 'AAG'),
        'L': ('TTA', 'TTG', 'CTT', 'CTC', 'CTA', 'CTG'),
        'M': ('ATG',),
        'N': ('AAT', 'AAC'),
        'P': ('CCT', 'CCC', 'CCA', 'CCG'),
        'Q': ('CAA', 'CAG'),
        'R': ('CGT', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG'),
        'S': ('TCT', 'TCC', 'TCA', 'TCG', 'AGT', 'AGC'),
        'T': ('ACT', 'ACC', 'ACA', 'ACG'),
        'V': ('GTT', 'GTC', 'GTA', 'GTG'),
        'W': ('TGG',),
        'Y': ('TAT', 'TAC')
    }
```

```python
f = SeqIO.parse(align, 'fasta')
codons = []
with open('seq_codon.csv', 'w') as file:   
    for row in f:
        codons = [str(row.seq[i:i+3]) for i in range(0, len(row.seq), 3)]
        keys = []
        for c in codons:
            keys += [key for key, value in codon_table.items() if c in value]
        dkeys = list(set(keys))
        d = {}
        for aa in dkeys:
            freq = {}
            for nucs in codon_table[aa]:
                freq[nucs] = codons.count(nucs)/keys.count(aa)
            d[aa] = freq
        for k,v in d.items():
            file.write(row.id+"," + k+","+str(v)+'\n')
```

```python
f = SeqIO.parse(align, 'fasta')
with open('codon_freq.csv', 'w') as file:
    for row in f:
        codons = [str(row.seq[i:i+3]) for i in range(0, len(row.seq), 3)]
      #  d = dict([(key, len(list(group))/len(codons)) for key, group in groupby(codons)])
        c = Counter(codons)
        for k, v in c.items():
            if "-" not in k:
                file.write(row.id + "," + k + ", " + str(v/len(codons)) + '\n')
```
